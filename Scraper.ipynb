```python
#2.GET LINKS FOR PHRASE AND MOVE TO CSV
def getlinks(driver):
    from selenium.common.exceptions import NoSuchElementException
    books = []
    check = "book/"
    import time
    for i in range (50):
        print(i)
        time.sleep(2)
        elems = driver.find_elements_by_xpath("//a[@href]")
        i = i+1
        time.sleep(3)
        for elem in elems:
            if (check in elem.get_attribute("href")):
                #print (elem.get_attribute("href"))
                books.append(elem.get_attribute("href"))
            else:
                pass
        try:
            time.sleep(1.5)
            driver.find_element_by_xpath('//*[@id="pagination1"]/div/ul/li[5]/a').click()
            time.sleep(2.5)
        except NoSuchElementException:  #Brak elementu 
            continue

    import pandas
    df = pandas.DataFrame(books)
    df = df.drop_duplicates([0], keep = 'first')
    df.to_csv("./cloud.csv", sep=',',index=False)
    df.shape
    
    return df
getlinks(driver)
    
```
